```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
imagem_bab = cv2.imread('dataset/city.png',cv2.IMREAD_GRAYSCALE)
```

```python
#Negativo

imagem_negativa = 255 - imagem_bab

plt.imshow(imagem_negativa, cmap='gray')
```

```python
# Intervalo para [100,200]

imagem_intervalo = imagem_bab.copy()
imagem_intervalo[imagem_intervalo > 200] = 200
imagem_intervalo[imagem_intervalo < 100] = 100

plt.imshow(imagem_intervalo, cmap='gray')
```

```python
#Inverter pares

imagem_invertida = imagem_bab.copy()

imagem_invertida[::2,0:] = imagem_bab[::2,::-1]

plt.imshow(imagem_invertida, cmap="gray")
```

```python
# Reflexão linhas

imagem_refletida = imagem_bab.copy()

altura, largura = imagem_refletida.shape

imagem_refletida[altura//2:altura, 0:] = imagem_refletida[altura//2:0:-1, 0:]

plt.imshow(imagem_refletida, cmap="gray")
```

```python
#Espelhamento vertical
altura, largura = imagem_bab.shape

imagem_espelhada = np.zeros((altura,largura))

imagem_espelhada[0:altura:1,0::] = imagem_bab[::-1,0::]

plt.imshow(imagem_espelhada, cmap="gray")
```
