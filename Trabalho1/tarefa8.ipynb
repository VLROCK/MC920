```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
imagem_pep = cv2.imread('dataset/peppers.png')
```

```python
copia = imagem_pep.copy().astype(np.float64)
# pode estar tendo um overflow que o ngc nn corrije
imagem_pep[:,:,0] = 0.393 * copia[:,:,0] + 0.769 * copia[:,:,1] + 0.189 * copia[:,:,2]
imagem_pep[:,:,1] = 0.349 * copia[:,:,0] + 0.686 * copia[:,:,1] + 0.168 * copia[:,:,2]
imagem_pep[:,:,2] = 0.272 * copia[:,:,0] + 0.534 * copia[:,:,1] + 0.131 * copia[:,:,2]

imagem_pep[imagem_pep>255] = 255

plt.imshow(imagem_pep)
```
