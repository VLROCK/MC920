```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
imagem1 = cv2.imread('dataset/watch.png')
imagem2 = cv2.imread('dataset/blue_velvet.png')
imagem3 = cv2.imread('dataset/after_hours.png')

imagem = imagem2
```

```python
imagem = imagem.astype(np.float64) / 255
copia = imagem.copy().astype(np.float64)

imagem[:,:,2] = 0.393 * copia[:,:,2] + 0.769 * copia[:,:,1] + 0.189 * copia[:,:,0]  # R'
imagem[:,:,1] = 0.349 * copia[:,:,2] + 0.686 * copia[:,:,1] + 0.168 * copia[:,:,0]  # G'
imagem[:,:,0] = 0.272 * copia[:,:,2] + 0.534 * copia[:,:,1] + 0.131 * copia[:,:,0]  # B'

imagem = np.clip(imagem,0,1)

imagem = imagem * 255

imagem = cv2.cvtColor(imagem.astype(np.uint8), cv2.COLOR_BGR2RGB)

plt.imshow(imagem)
#plt.imsave('imagens_transformadas/tarefa8/after_hours_s.png',imagem)
```
