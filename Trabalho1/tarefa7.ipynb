```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
imagem1 = cv2.imread('dataset/watch.png')
imagem2 = cv2.imread('dataset/blue_velvet.png')
imagem3 = cv2.imread('dataset/after_hours.png')

imagem = imagem2
```

```python
plt.imshow(imagem)
```

```python
matriz =[
[0.393, 0.769, 0.189],
[0.349, 0.686, 0.168],
[0.272, 0.534, 0.131]
]

copia_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB).astype(np.float64)

imagem_filtro = np.dot(copia_rgb, np.array(matriz).T)

imagem_filtro = np.clip(imagem_filtro, 0, 255).astype(np.uint8)

plt.imshow(imagem_filtro)
#plt.imsave('imagens_transformadas/tarefa7/after_hours.png', imagem_filtro)
```
