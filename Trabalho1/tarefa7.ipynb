```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
imagem = cv2.imread('dataset/watch.png')
```

```python
matriz =[
[0.393, 0.769, 0.189],
[0.349, 0.686, 0.168],
[0.272, 0.534, 0.131]
]

copia = imagem.copy().astype(np.float64)

imagem_filtro = np.dot(copia, np.array(matriz).T)

imagem_filtro = np.clip(imagem_filtro,0,255).astype(np.uint8)



plt.imshow(imagem_filtro)
```
