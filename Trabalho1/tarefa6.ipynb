```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
imagem = cv2.imread('dataset/baboon_monocromatica.png',cv2.IMREAD_GRAYSCALE)
```

```python
nova_ordem = np.array([
    [ 6, 11, 13,  3],
    [ 8, 16,  1,  9],
    [12, 14,  2,  7],
    [ 4, 15, 10,  5]
]) - 1 
```

```python
altura, largura = imagem.shape

tamanho_bloco = altura // 4

lista_blocos = imagem.reshape(4,tamanho_bloco,4,tamanho_bloco)

plt.imshow(lista_blocos[0,::,0,::], cmap="grey")

lista_blocos = lista_blocos.transpose(0,2,1,3)

lista_blocos = lista_blocos.reshape(16,tamanho_bloco, tamanho_bloco)
```

```python
imagem_mosaico = lista_blocos[nova_ordem]

imagem_mosaico = imagem_mosaico.reshape(4,4,tamanho_bloco,tamanho_bloco)

imagem_mosaico = imagem_mosaico.transpose(0,2,1,3)

imagem_mosaico = imagem_mosaico.reshape(altura, largura)

plt.imshow(imagem_mosaico, cmap="grey")
```
