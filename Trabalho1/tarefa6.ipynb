```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
imagem1 = cv2.imread('dataset/baboon_monocromatica.png',cv2.IMREAD_GRAYSCALE)
imagem2 = cv2.imread('dataset/eraserhead4.png',cv2.IMREAD_GRAYSCALE)
imagem3 = cv2.imread('dataset/o_farol.png',cv2.IMREAD_GRAYSCALE)
imagem4 = cv2.imread('dataset/seagull.png',cv2.IMREAD_GRAYSCALE)

imagem = imagem2
```

```python
nova_ordem = np.array([
    [ 6, 11, 13,  3],
    [ 8, 16,  1,  9],
    [12, 14,  2,  7],
    [ 4, 15, 10,  5]
]) - 1 
```

```python
plt.imshow(imagem, cmap='gray')
```

```python
altura, largura = imagem.shape

bloco_h = altura // 4
bloco_w = largura // 4

lista_blocos = imagem.reshape(4, bloco_h, 4, bloco_w)


plt.imshow(lista_blocos[0,::,0,::], cmap="grey")

lista_blocos = lista_blocos.transpose(0,2,1,3)

lista_blocos = lista_blocos.reshape(16,bloco_h, bloco_w)
```

```python
imagem_mosaico = lista_blocos[nova_ordem]

imagem_mosaico = imagem_mosaico.reshape(4,4,bloco_h,bloco_w)

imagem_mosaico = imagem_mosaico.transpose(0,2,1,3)

imagem_mosaico = imagem_mosaico.reshape(altura, largura)

plt.imshow(imagem_mosaico, cmap="grey")
plt.imsave('imagens_transformadas/tarefa6/eraserhead.png', imagem_mosaico, cmap='gray')
```
