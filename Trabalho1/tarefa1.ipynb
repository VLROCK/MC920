```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

OBSERVAÇÃO!!

Para testar com uma imagem diferente, adicione o caminho da imagem abaixo e atribua à variável imagem:

imagem4 = cv2.imread('dataset/nova_imagem.png')

imagem = imagem4

Desse modo, todas as células subsequentes processarão automaticamente a nova 
imagem, sem necessidade de qualquer outra alteração no código.


Outros casos de testes podem ser encontrados em imagens_transformada/tarefaX.

Os plt.imsave() estã comentados para não alterar os arquivos de testes.

```python
imagem1 = cv2.imread('dataset/baboon_monocromatica.png',cv2.IMREAD_GRAYSCALE)

imagem2 = cv2.imread('dataset/seagull.png', cv2.IMREAD_GRAYSCALE)

imagem3 = cv2.imread('dataset/o_farol.png', cv2.IMREAD_GRAYSCALE)

imagem = imagem3
```

```python
#Imagem original

plt.imshow(imagem, cmap='gray')
```

Observação: cada rotação possui 3 casos de testes que estão salvos em imagens_transformadas/tarefa1.

```python
#Rotação em 90 graus
altura, largura = imagem.shape

imagem_90 = np.zeros((largura, altura))
i = np.arange(0,altura,1)
j = np.arange(0,largura,1)
grade = np.ix_(i, j)

imagem_90[grade[1], (altura -1) - grade[0]] = imagem[grade[0], grade[1]]

plt.imshow(imagem_90, cmap='gray')
#plt.imsave('imagens_transformadas/tarefa1/f_resultado_90.png', imagem_90, cmap='gray')
```

```python
#Rotação em 180 graus
altura, largura = imagem.shape

imagem_180 = np.zeros((altura, largura))
i = np.arange(0,altura,1)
j = np.arange(0,largura,1)
grade = np.ix_(i, j)

imagem_180[(altura - 1) - grade[0], (largura - 1) - grade[1]] = imagem[grade[0], grade[1]]

plt.imshow(imagem_180, cmap='gray')
#plt.imsave('imagens_transformadas/tarefa1/f_resultado_180.png', imagem_180, cmap='gray')
```

```python
#Rotação em 270 graus
altura, largura = imagem.shape

imagem_270 = np.zeros((largura, altura))
i = np.arange(0,altura,1)
j = np.arange(0,largura,1)
grade = np.ix_(i, j)

imagem_270[(largura - 1) - grade[1], grade[0]] = imagem[grade[0], grade[1]]

plt.imshow(imagem_270, cmap='gray')
#plt.imsave('imagens_transformadas/tarefa1/f_resultado_270.png', imagem_270, cmap='gray')
```
