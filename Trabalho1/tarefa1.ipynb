```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
imagem = cv2.imread('dataset/baboon_monocromatica.png',cv2.IMREAD_GRAYSCALE)
```

```python
altura, largura = imagem.shape

imagem_90 = np.zeros((largura, altura))
i = np.arange(0,largura,1)
j = np.arange(0,altura,1)
grade = np.ix_(i, j)

imagem_90[grade[1], (altura -1) - grade[0]] = imagem[grade[0], grade[1]]

plt.imshow(imagem_90, cmap='gray')
```

```python
altura, largura = imagem.shape

imagem_180 = np.zeros((altura, largura))
i = np.arange(0,largura,1)
j = np.arange(0,altura,1)
grade = np.ix_(i, j)

imagem_180[(altura - 1) - grade[0], (largura - 1) - grade[1]] = imagem[grade[0], grade[1]]

plt.imshow(imagem_180, cmap='gray')
```
