```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
imagem1 = cv2.imread('dataset/baboon_monocromatica.png',cv2.IMREAD_GRAYSCALE)
imagem2 = cv2.imread('dataset/o_farol.png',cv2.IMREAD_GRAYSCALE)
imagem3 = cv2.imread('dataset/eraserhead.png',cv2.IMREAD_GRAYSCALE)
```

```python
imagem = np.copy(imagem3)
imagem_normalizada = imagem / 255
g = 3.5


imagem_nova = imagem_normalizada ** (1/g)

imagem_nova = np.round(imagem_nova * 255).astype(np.uint8)

plt.imshow(imagem_nova, cmap='gray')
#plt.imsave(f'imagens_transformadas/tarefa4/eraserhead_{g}.png',imagem_nova, cmap='gray')
```
