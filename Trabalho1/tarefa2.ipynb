```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

OBSERVAÇÃO!!

Para testar com uma imagem diferente, adicione o caminho da imagem abaixo e atribua à variável imagem:

imagem4 = cv2.imread('dataset/nova_imagem.png')

imagem = imagem4

Desse modo, todas as células subsequentes processarão automaticamente a nova 
imagem, sem necessidade de qualquer outra alteração no código.


Outros casos de testes podem ser encontrados em imagens_transformada/tarefaX.

Os plt.imsave() estã comentados para não alterar os arquivos de testes.

```python
imagem1= cv2.imread('dataset/baboon_monocromatica.png',cv2.IMREAD_GRAYSCALE)
imagem2 = cv2.imread('dataset/eraserhead.png',cv2.IMREAD_GRAYSCALE)


imagem = imagem2
```

```python
plt.imshow(imagem, cmap='gray')
print(imagem.shape)
```

```python
altura, largura = imagem.shape

imagem_2x = np.zeros((altura*2, largura*2))

idx_a = np.arange(altura*2)
idx_l = np.arange(largura*2)
grade = np.ix_(idx_a, idx_l)


imagem_2x[grade[0], grade[1]] = imagem[grade[0] // 2, grade[1] // 2]

plt.imshow(imagem_2x, cmap='gray')
#plt.imsave('imagens_transformadas/tarefa2/2xeraserhead.png', imagem_2x, cmap='gray')
```

```python
altura, largura = imagem.shape

imagem_4x = np.zeros((altura*4, largura*4))

idx_a = np.arange(altura*4)
idx_l = np.arange(largura*4)
grade = np.ix_(idx_a, idx_l)


imagem_4x[grade[0],grade[1]] = imagem[grade[0] // 4, grade[1] // 4]

plt.imshow(imagem_4x, cmap='gray')
#plt.imsave('imagens_transformadas/tarefa2/4xeraserhead.png', imagem_4x, cmap='gray')
```

```python
altura, largura = imagem.shape

n = 3

imagem_nx = np.zeros((altura*n, largura*n))

idx_a = np.arange(altura*n)
idx_l = np.arange(largura*n)
grade = np.ix_(idx_a, idx_l)


imagem_nx[grade[0],grade[1]] = imagem[grade[0] // n, grade[1] // n]

plt.imshow(imagem_nx, cmap='gray')
#plt.imsave('imagens_transformadas/tarefa2/nxeraserhead.png', imagem_nx, cmap='gray')
```
