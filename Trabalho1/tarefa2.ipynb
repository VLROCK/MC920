```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
imagem_mono_bab = cv2.imread('dataset/baboon_monocromatica.png',cv2.IMREAD_GRAYSCALE)
```

```python
largura, altura = imagem_mono_bab.shape

imagem_2x = np.zeros((largura*2, altura*2))

idx_l = np.arange(largura*2)
idx_a = np.arange(altura*2)
grade = np.ix_(idx_l, idx_a)


imagem_2x[grade[0], grade[1]] = imagem_mono_bab[grade[0] // 2, grade[1] // 2]

plt.imshow(imagem_2x, cmap='gray')
```

```python
largura, altura = imagem_mono_bab.shape

imagem_4x = np.zeros((largura*4, altura*4))

idx_l = np.arange(largura*4)
idx_a = np.arange(altura*4)
grade = np.ix_(idx_l, idx_a)


imagem_4x[grade[0],grade[1]] = imagem_mono_bab[grade[0] // 4, grade[1] // 4]

plt.imshow(imagem_4x, cmap='gray')
```
