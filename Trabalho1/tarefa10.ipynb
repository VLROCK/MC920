```python
imagem_bab = cv2.imread('dataset/baboon_monocromatica.png',cv2.IMREAD_GRAYSCALE)

imagem_bor = cv2.imread('dataset/butterfly.png',cv2.IMREAD_GRAYSCALE)

imagem_sea = cv2.imread('dataset/seagull.png',cv2.IMREAD_GRAYSCALE)

```

```python
x = 0.5
imagem_nova = imagem_bab*x + imagem_sea*(1-x)
imagem_nova[imagem_nova > 255] = 255

plt.imshow(imagem_nova, cmap='gray')
#plt.imsave(f'imagens_transformadas/tarefa10/imagem_{x}.png', imagem_nova, cmap='gray')
```
